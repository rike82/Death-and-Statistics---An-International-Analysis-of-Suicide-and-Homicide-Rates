```python
def show_details(df):
    pd.set_option('display.max_columns', None)  # for showing all the columns of a df
   #print(f"HEAD OF DF:\n{df.head()}\n\n")
    print("DF-INFO:\n")
    df.info()
    print(      
        f"\nALL COLUMNS:\n {df.columns}\n\n"
        f"DESCRIPTION OF DF:\n{df.describe()}\n\n"
        f"DF-SHAPE: {df.shape}\n\n"
        f"COUNT OF UNIQUE VALUES:\n {df.nunique()}\n\n"
        f"DF-DATATYPES:\n{df.dtypes}"    
    )
```

```python
def load_file(path):
    # check if it is a csv
    if str(path).split('.')[-1] == "csv":
        df = pd.read_csv(path)
    # check if it is excel
    elif str(path).split('.')[-1] in ["xls", "xlsx"]:
        df = pd.read_excel(path)
    else:
        raise ValueError("Unsupported file type. Please use a CSV or Excel file.")
    
    return df
```

```python
def year_dfs(df, year, columns):
    # filter df for certain year
    df_year = df[df['year'] == year]
    print(f"\nYear: {year}")
    # iterates over columns and prints unique values for them
    for col in columns:
        print(f"Unique values in {col}: {df_year[col].unique()}")
```

<h3>Importing and exploring/cleaning of Data - Homicide Rates (Male, Female, ALL Genders)</h3>

```python
df6_male = load_file('..\\Homicide\\Killed_male.csv')
df6_male.head()
```

```python
df6_male.drop(columns = ['Series Name','Series Code'], inplace= True)
df6_male.rename(columns = {'Country Code' : 'iso3c', 'Country Name' : 'country'}, inplace= True)
df6_male.head()
```

```python
# use mapping on df6_male codes, set country names to names in mapping
df6_male['Mapped_Country'] = df6_male['iso3c'].map(country_mapping).fillna('NotFoundInMapping')
df6_male['Mapped_Country'].nunique()
```

```python
df6_male = drop_not_mapped(df6_male)
df6_male.head()
```

```python
def easy_pivoting(df, columns_in, new_column):
    # melting df to show years in 'year' column
    df_melted = pd.melt(
        df,
        id_vars=columns_in,  # keep columns 'country' and 'iso3c'
        var_name='year',     # put all years in column 'year' 
        value_name=new_column  # values from year-columns appear under'homicide_male'
    )
    
    # extract only numbers of the year columns (2010 [YR2010])
    df_melted['year'] = df_melted['year'].str.extract(r'(\d{4})')[0]
    
    return df_melted


```

```python
# example for calling the method
cols = ['country', 'iso3c']
new_column = 'homicide_male'
df6_male = easy_pivoting(df6_male, cols, new_column)

# show transformed dataframe
df6_male.head(30)
```

```python
df6_female = load_file('..\\Homicide\\Killed_female.csv')
df6_all = load_file('..\\Homicide\\All.csv')
```

```python
def easy_clean_and_map(df):
    # drop and rename as in male_Data
    df.drop(columns = ['Series Name','Series Code'], inplace= True)
    df.rename(columns = {'Country Code' : 'iso3c', 'Country Name' : 'country'}, inplace= True)
    #set country names to names in mapping
    df['Mapped_Country'] = df['iso3c'].map(country_mapping).fillna('NotFoundInMapping')
    return df
```

```python
df6_female = easy_clean_and_map(df6_female)
df6_female = drop_not_mapped(df6_female)
df6_female
```

```python
cols = ['country', 'iso3c']
new_column = 'homicide_female'
df6_female = easy_pivoting(df6_female, cols, new_column)
df6_female.head(30)
```

```python
df6_all = easy_clean_and_map(df6_all)
df6_all = drop_not_mapped(df6_all)
df6_all
```

```python
cols = ['country', 'iso3c']
new_column = 'homicide_all'
df6_all = easy_pivoting(df6_all, cols, new_column)
df6_all.head(30)
```

```python
# first merge with homicide_female
df_combined = pd.merge(df6_male, df6_female[['country', 'iso3c', 'year', 'homicide_female']], 
                       on=['country', 'iso3c', 'year'], how='left')

# merge again to add homicide_all
df_combined = pd.merge(df_combined, df6_all[['country', 'iso3c', 'year', 'homicide_all']], 
                       on=['country', 'iso3c', 'year'], how='left')
```

```python
# convert 'year' to int
df_combined['year'] = pd.to_numeric(df_combined['year'], errors='coerce').astype('Int64')  # 'Int64' NaN-values are possible

# convert 'homicide_male', 'homicide_female', 'homicide_all' to float
for col in ['homicide_male', 'homicide_female', 'homicide_all']:
    df_combined[col] = pd.to_numeric(df_combined[col], errors='coerce')
```

```python
# clean column names
df7_filtered.columns = df7_filtered.columns.str.strip()
# replace '-' with NaN
df7_filtered.replace('-', pd.NA, inplace=True)
# melt year-columns to column 'year'
df_melted = pd.melt(df7_filtered, 
                    id_vars=['iso3c', 'country', 'Indicator ID', 'Indicator', 'Attribute 1'], 
                    var_name='year', 
                    value_name='value')

# pivot, to have'Indicator' as separate column, with'value' their related values
df_pivoted = df_melted.pivot_table(index=['iso3c', 'country', 'year'], 
                                   columns=['Indicator', 'Attribute 1'], 
                                   values='value',
                                   aggfunc='first')

# names for the multi-attribute columns
df_pivoted.columns = ['_'.join(col).strip() for col in df_pivoted.columns.values]

# always reset index
df_pivoted.reset_index(inplace=True)

df_pivoted.head()
```

```python
# shorten column names!!
df_pivoted.rename(columns= {'Global Gender Gap Economic Participation and Opportunity Subindex_Index' : 'Econ_Particip_and_Opport_Subind_Index'}, inplace= True)
df_pivoted.rename(columns= {'Global Gender Gap Economic Participation and Opportunity Subindex_Rank' : 'Econ_Particip_and_Opport_Subind_Rank'}, inplace= True)

df_pivoted.rename(columns= {'Global Gender Gap Educational Attainment Subindex_Index' : 'Educ_Attainmnt_Subind_Index'}, inplace= True)
df_pivoted.rename(columns= {'Global Gender Gap Educational Attainment Subindex_Rank' : 'Educ_Attainmnt_Subind_Rank'}, inplace= True)


df_pivoted.rename(columns= {'Global Gender Gap Health and Survival Subindex_Index' : 'Health_and_Survival_Subind_Index'}, inplace= True)
df_pivoted.rename(columns= {'Global Gender Gap Health and Survival Subindex_Rank' : 'Health_and_Survival_Subind_Rank'}, inplace= True)



df_pivoted.rename(columns= {'Global Gender Gap Political Empowerment subindex_Index' : 'Polit_Empowmnt_Subind_Index'}, inplace= True)
df_pivoted.rename(columns= {'Global Gender Gap Political Empowerment subindex_Rank' : 'Polit_Empowmnt_Subind_Rank'}, inplace= True)

df_pivoted.rename(columns= {'Overall Global Gender Gap Index_Index' : 'Overall_GGGap_Ind_Index'}, inplace= True)
df_pivoted.rename(columns= {'Overall Global Gender Gap Index_Rank' : 'Overall_GGGap_Ind_Rank'}, inplace= True)


```

```python
df_pivoted.head()
```

```python
df_pivoted.dtypes
```

```python
# Convert 'year' column to integers
df_pivoted['year'] = df_pivoted['year'].astype(int)

# Convert indicator columns to floats (assuming all have '_Index' or similar suffix for numeric values)
indicator_cols = [col for col in df_pivoted.columns if '_Index' in col]
rank_cols = [col for col in df_pivoted.columns if '_Rank' in col]

# Apply conversion to float for indicator columns
df_pivoted[indicator_cols] = df_pivoted[indicator_cols].apply(pd.to_numeric, errors='coerce')

# Convert rank columns to integer, with support for NaN values
df_pivoted[rank_cols] = df_pivoted[rank_cols].apply(pd.to_numeric, errors='coerce').astype('Int64')

# Check data types to confirm the change
print(df_pivoted[indicator_cols].dtypes)
print(df_pivoted[rank_cols].dtypes)
print(df_pivoted['year'].dtypes)
```
