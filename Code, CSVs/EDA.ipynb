```python
df_long = df.melt(
    id_vars=["year", "country"],  # keeps these columns
    value_vars=[
        "TOTAL_Y15T24", "TOTAL_Y25T34", "TOTAL_Y35T44", "TOTAL_Y45T54",
        "TOTAL_Y55T64", "TOTAL_Y65T74", "TOTAL_Y75T84", "TOTAL_Y_GE85"
    ],
    var_name="age_group",  # new column for age_groups
    value_name="suicide_rate"  # new column for suicide rates
)

df_long
```

```python
# filter for ten years 2011-2020
df_filtered = df_long[df_long["year"].between(2011, 2021)]

# calculate the average
df_avg = (
    df_filtered
    .groupby(["country", "age_group"], as_index=False)
    .agg({"suicide_rate": "mean"})
)

df_avg
```

```python
df_avg.to_csv('age_groups_countries.csv')
```
