```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('Hom_sui_GDP.csv')
df1.drop(columns = ['Unnamed: 0'], inplace= True)
df1.dtypes
```

```python
# suicide rates age_standardized
df2 = pd.read_csv('agestandardized_suicide.csv')
df2.drop(columns = ['Unnamed: 0'], inplace= True)
df2
```

```python
df3 = pd.read_csv('WorldPop_density.csv')
df3.drop(columns = ['Unnamed: 0'], inplace= True)
df3.head()
```

```python
df4 = pd.read_csv('WorldPop_lifequality.csv')
df4.drop(columns = ['Unnamed: 0'], inplace= True)
df4.head()
```

```python
df5 = pd.read_csv('WorldPop_educationAndTemp.csv')
df5.drop(columns = ['Unnamed: 0'], inplace= True)
df5.head()
```

```python
df6 = pd.read_csv('homicide_by_gender.csv')
df6.drop(columns = ['Unnamed: 0'], inplace= True)
df6.head()
```

```python
df7 = pd.read_csv('gender_gap_data.csv')
df7.drop(columns = ['Unnamed: 0'], inplace= True)
df7.head()
```

```python
df8 = pd.read_csv('clean_income_dist.csv')
df8.drop(columns = ['Unnamed: 0'], inplace= True)
df8.head()
```

```python
#remove duplicates in df1 and df2 in columns 'iso3c' and 'country'
df1_selected = df1[['iso3c', 'country', 'adminregion']].drop_duplicates(subset=['iso3c', 'country'])
df2_selected = df2[['iso3c', 'country', 'ParentLocationCode', 'ParentLocation']].drop_duplicates(subset=['iso3c', 'country'])

# merge dataframes
df_combined = pd.merge(df1_selected, df2_selected, on=['iso3c', 'country'], how='left')


```

```python
df_combined
```

```python
df_combined2 = pd.merge(df_combined, df3[['country', 'area', 'population', 'pop_per_km_sq']], 
                       on=['country'], how='left')

df_combined2
```

```python
df_combined3 = pd.merge(df_combined2, df4[['country', 'stability', 'rights', 'health', 'safety','costs']], 
                       on=['country'], how='left')

df_combined3   

```

```python
df_combined4 = pd.merge(df_combined3, df5[['country', 'iq', 'education_expenditure_per_inhabitant', 'daily_max_temp']], 
                       on=['country'], how='left')

df_combined4 
```

```python
df_combined4.dtypes
```

```python
df_combined4.to_csv('country_table1.csv', index=False)
```
